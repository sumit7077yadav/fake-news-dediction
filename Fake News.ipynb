# --> Drop both columns

```python
df.drop(columns=['news_url','tweet_num'],axis=1,inplace=True)
```

```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
stop_words = set(stopwords.words('english'))
len(stop_words)
```

```python
def remove_stopwords(text):
    words=word_tokenize(text)
    return ' '.join([word for word in words if word.lower() not in stop_words])
```

```python
df['title'] = df['title'].apply(remove_stopwords)
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
def clean_title(line):
    line = str(line)
    line = re.sub(r'[^a-zA-Z\s]', '' ,line)
    line = line.lower()
    return line
```

```python
df['title'] = df['title'].apply(clean_title)
```

```python
df['source_domain'] = df['source_domain'].apply(clean_title)
```

```python
df['text'] = df['title'] + ' ' + df['source_domain']
```

```python
df.head()
```

```python
x= df['text'].values
y = df['real'].values
```

# --> Using Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

# --> Using TfidVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import sklearn
```

```python
x_train , x_test ,y_train , y_test = train_test_split(x,y ,test_size=1/3,random_state=42)
```

```python
vectorizer = TfidfVectorizer()
```

```python
x_train = vectorizer.fit_transform(x_train)
x_test = vectorizer.transform(x_test)
```

# --> Using LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
pred = logistic.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```

# --> Using DecisionTree

```python
from sklearn import tree
```

```python
tree1 = tree.DecisionTreeClassifier()
```

```python
tree1.fit(x_train,y_train)
```

```python
pred1 = tree1.predict(x_test)
```

```python
accuracy_score(pred1,y_test)
```

# --> Using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Random = RandomForestClassifier()
```

```python
Random.fit(x_train,y_train)
```

```python
pred2 = Random.predict(x_test)
```

```python
accuracy_score(pred2,y_test)
```

# --> Using XGBoost

```python
import xgboost
```

```python
xgb = xgboost.XGBClassifier()
```

```python
xgb.fit(x_train,y_train)
```

```python
pred3 = xgb.predict(x_test)
```

```python
accuracy_score(pred3,y_test)
```
